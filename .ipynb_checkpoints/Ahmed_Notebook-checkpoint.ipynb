```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
ppd_df = pd.read_csv("ppd_complaints.csv")
```

```python
display(ppd_df.head())
```

```python
# Add date of complaint
date_pattern = r"(\d{1,2}-\d{1,2}-\d{1,4})" #parentheses is to treat all characters as a "group", making extract work
complaint_str_series = ppd_df['summary'].str.extract(date_pattern, expand=False) #expand=False to get series
complaint_str_series.replace("2-29-18","03-01-18",inplace=True) #replace faulty leap year day
ppd_df['date_complaint_made'] = pd.to_datetime(complaint_str_series)
```

```python
# Basic information and visualizations
print("Number of complaints: {}".format(ppd_df.shape[0]))

print("Complaints over time")
# optionally, exclude the first few years
complaints_over_time_series = ppd_df[ppd_df['date_complaint_made']>"01-01-2014"].groupby('date_complaint_made').size()
#complaints_over_time_series = ppd_df.groupby('date_complaint_made').size().asfreq(freq="d",fill_value=0)
display(complaints_over_time_series.groupby(pd.Grouper(freq="M")).sum().plot(figsize=(20,10)))
```
